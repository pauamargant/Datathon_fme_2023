```python
import pandas as pd
import pickle
import torch  
```

```python
product_data = pd.read_csv('dataset/product_data_preprocessed.csv')
product_data.head()
```

```python
with open('dataset/compatibilities.pickle', 'rb') as f:
    compatibilities = pickle.load(f)

labels = pd.DataFrame.from_dict(compatibilities.items())
labels.columns = ['cod_modelo_color', 'community_id']
product_data = labels.merge(product_data, on='cod_modelo_color')
print(product_data.shape)
product_data.head()
```

```python
with open('../image_features.pkl', 'rb') as f:
    image_features = pickle.load(f)

product_data['des_filename'] = product_data['des_filename'].apply(lambda x: x.split('/')[-1])

image_features.rename(columns={'Filename':'des_filename'}, inplace=True)
product_data = product_data.merge(image_features, on = 'des_filename').drop(['des_filename'], axis=1)
```

```python
dict_modelo_index = dict(zip(product_data.cod_modelo_color, product_data.index))

dict_index_modelo = dict(zip(product_data.index, product_data.cod_modelo_color))
# save as dict index_modelo.pickle
with open('dataset/dict_index_modelo.pickle', 'wb') as f:
    pickle.dump(dict_index_modelo, f, pickle.HIGHEST_PROTOCOL)
# create cod_modelo_color : index dict
# save as dict modelo_index.pickle
with open('dataset/dict_modelo_index.pickle', 'wb') as f:
    pickle.dump(dict_modelo_index, f, pickle.HIGHEST_PROTOCOL)
X = product_data.drop(['community_id', 'cod_modelo_color'], axis=1)
y = product_data['community_id']
```
